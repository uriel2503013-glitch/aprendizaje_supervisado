# 1 Limpieza de Datos

```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('titanic_clean.csv')
```

```python
df.head()
```

```python
print("valores nulos por columna")
print(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

df_clean = df.drop(['Cabin', 'Ticket', 'Name', 'PassengerId', 'Parch'], axis=1)

df_clean.to_csv('titanic_clean.csv', index=False)
print("Datos guardados en titanic_clean.csv")
```
